# Python JSON Read & Write

Python provides a built-in module called json to work with JSON data.

# 1. Import JSON Module

```python
import json
```

# üéØ 2. Writing JSON to a File (json.dump)
**‚úî Write Python dictionary ‚Üí JSON file**

```python
import json

data = {
    "name": "Santhosh",
    "age": 25,
    "city": "Chennai"
}

with open("data.json", "w") as f:
    json.dump(data, f, indent=4)
```

# üìå Notes:

**json.dump() **‚Üí writes to a file

**indent=4** ‚Üí makes file readable (pretty print)

# 3. Reading JSON From a File (json.load)

```python
import json

with open("data.json", "r") as f:
    data = json.load(f)

print(data)
print(data["name"])
```

**üìå Notes:**

json.load() ‚Üí reads from a file

Converts JSON ‚Üí Python dictionary

# üéØ 4. Converting Python Data to JSON String (json.dumps)

```python
person = {"name": "Kumar", "age": 30}

json_string = json.dumps(person)
print(json_string)
```

**üìå Output:**

```python
{"name": "Kumar", "age": 30}
```

# 5. Converting JSON String to Python Data (json.loads)

```python
json_data = '{"name": "Ravi", "age": 22}'

py_data = json.loads(json_data)
print(py_data)
```

üìå Output:**bold text**

```python
{'name': 'Ravi', 'age': 22}
```

# 6. JSON Supported Python Data Types

| None        | null            |


# 7. Handling Errors in JSON

‚ùó Invalid JSON

```python
try:
    data = json.loads("{name:}")
except json.JSONDecodeError:
    print("Invalid JSON format")
```

# 8. Examples of Complex JSON

**Write list of dictionaries to JSON**

```python
students = [
    {"id": 1, "name": "Arun"},
    {"id": 2, "name": "Bala"}
]

with open("students.json", "w") as f:
    json.dump(students, f, indent=4)
```
