#  1. What is a Function?

A function is a block of code that runs only when called.

# 2. Function Without Parameters

```python
def greet():
    print("Hello! Welcome to Python")

greet()
```

# 3. Function With parameters

```python
def add(a, b):
    print("Sum =", a + b)

add(5, 3)
```

# 4. Function With Return Value

```python
def square(n):
    return n * n

result = square(4)
print(result)
```

# 5. Default Parameter

```python
def welcome(name="Guest"):
    print("Hello", name)

welcome()
welcome("Santhosh")
```

# 6. Function With Multiple Returns

```python
def calc(a, b):
    return a+b, a-b, a*b

x, y, z = calc(10, 5)
print(x, y, z)
```

# 7. Variable-length Arguments

args ‚Üí accepts any number of values

```python
def total(*nums):
    return sum(nums)

print(total(10, 20, 30))
```

# kwargs ‚Üí accepts key-value pairs

```python
def show_details(**info):
    print(info)

show_details(name="Arun", age=20)
```

# ‚≠ê MINI PROGRAMS (VERY IMPORTANT)

üü¶ Mini Program 1: Find Factorial Using Function

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

# Mini Program 2: Check Even or Odd

```python
def check(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check(7))
```

# Mini Program 3: Count Vowels in a String

```python
def count_vowels(text):
    count = 0
    vowels = "aeiouAEIOU"
    for ch in text:
        if ch in vowels:
            count += 1
    return count

print(count_vowels("hello world"))
```

# Mini Program 4: Simple Calculator

```python
def calculator(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return a / b
    else:
        return "Invalid Operator"

print(calculator(10, 5, "*"))
```

# Mini Program 5: Largest of Three Numbers

```python
def largest(a, b, c):
    return max(a, b, c)

print(largest(10, 25, 6))
```

# Mini Program 6: Reverse a String

```python
def reverse_string(s):
    rev = ""
    for ch in s:
        rev = ch + rev
    return rev

print(reverse_string("python"))
```

# üü¶ Mini Program 7: Palindrome Check

```python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
```
