```python
üü™ TUPLE SOLUTIONS
1. Create a Tuple
fruits = ("apple", "banana", "cherry", "mango", "orange")

print(fruits[0])   # apple
print(fruits[-1])  # orange

2. Tuple Slicing
t = (10, 20, 30, 40, 50, 60)

print(t[:3])     # (10, 20, 30)
print(t[-3:])    # (40, 50, 60)
print(t[2:5])    # (30, 40, 50)

3. Count Occurrences
nums = (25, 10, 25, 30, 40, 25)
print(nums.count(25))   # 3

4. Tuple Concatenation
a = (1, 2, 3)
b = (4, 5, 6)

c = a + b
print(c)   # (1, 2, 3, 4, 5, 6)

5. Check Value
fruits = ("apple", "banana", "cherry")

print("banana" in fruits)   # True

6. Min & Max
t = (50, 20, 90, 10, 75)

print(min(t))   # 10
print(max(t))   # 90

üü© SET SOLUTIONS
7. Create a Set
s = {10, 20, 30, 40, 50}
print(s)

8. Add Items to Set
s = {1, 2, 3}

s.add(4)
s.update([5, 6, 7])

print(s)   # {1, 2, 3, 4, 5, 6, 7}

9. Remove Items Safely
s = {10, 20, 30}

s.remove(20)      # Removes 20
s.discard(30)     # Removes 30
s.discard(100)    # No error even if not present

print(s)          # {10}

10. Set Operations
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # Union ‚Üí {1, 2, 3, 4, 5, 6}
print(A & B)   # Intersection ‚Üí {3, 4}
print(A - B)   # Difference ‚Üí {1, 2}
print(A ^ B)   # Symmetric difference ‚Üí {1, 2, 5, 6}

11. Check Subset
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}

print(A.issubset(B))     # True
print(B.issuperset(A))   # True

12. Remove Duplicates from List
nums = [1, 2, 2, 3, 4, 4, 5]

unique_list = list(set(nums))
unique_list.sort()

print(unique_list)   # [1, 2, 3, 4, 5]

‚≠ê BONUS SOLUTIONS
13. Frequency Count Using Set
nums = [1, 2, 2, 3, 3, 3, 4]

unique = set(nums)

for n in unique:
    print(n, "‚Üí", nums.count(n))


Output:

1 ‚Üí 1
2 ‚Üí 2
3 ‚Üí 3
4 ‚Üí 1

14. Tuple of Sets
t = ({1,2}, {3,4}, {5,6})

# Convert to list because tuple items can't be replaced
temp = list(t)

temp[1].add(7)

t = tuple(temp)

print(t)


Output:

({1, 2}, {3, 4, 7}, {5, 6})

15. Frozenset Example
fs = frozenset([10, 20, 30, 40])
print(fs)

# Trying this will cause an error:
# fs.add(50)
```
