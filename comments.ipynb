# Single-Line Comments (The Hash Symbol #)

This is the most common and standard way to write comments in Python. Everything following the hash symbol (#) on that line is considered a comment and is ignored by the interpreter.

# As a Standalone Line
Use a full line for a comment to explain the code that follows it.

```python
# This is a comment. It explains the purpose of the next block of code.
def calculate_area(length, width):
    return length * width
```

# Inline with Code
Use a comment on the same line as a code statement, often to explain a specific detail or non-obvious choice. It is best practice to separate the code from the comment by at least two spaces.

# Multi-Line Comments (Docstrings)

Python does not have a specific syntax for multi-line block comments like some other languages (e.g., /* ... */). Instead, multi-line strings enclosed in triple quotes ("""...""" or '''...''') are used.

When these multi-line strings are placed as the first statement inside a module, class, or function, they are specifically called Docstrings and are used for documentation.

A. Docstrings (The Python Standard)
Docstrings are officially recognized by Python and are accessible at runtime using the __doc__ attribute or the built-in help() function.

```python
def greeting(name):
    """
    This function takes a name as input and returns a personalized greeting.

    It is intended for a single user interaction.
    """
    return f"Hello, {name}!"

# You can access this documentation in the console:
# >>> help(greeting)
```
