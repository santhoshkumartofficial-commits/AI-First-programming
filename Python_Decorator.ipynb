# 1. What is a Decorator?

A decorator is a function that takes another function and adds extra functionality to it without changing its original code.

‚úî It follows the concept of Higher-Order Functions
‚úî Often used for logging, authentication, timing, validation, etc.

# 2. Basic Decorator Structure

```python
def decorator_function(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper
```

```python
@decorator_function
def say_hello():
    print("Hello!")

say_hello()
```

# 3. How Decorators Work (Diagram)

```python
say_hello()
      ‚Üì
decorate ‚Üí wrapper() ‚Üí extra code + original code
```

Decorator replaces the original function with the wrapper function.

# 4. Decorator without @ Syntax

```python
def greeting():
    print("Hello!")

decorated = decorator_function(greeting)
decorated()
```

# 5. Decorator with Arguments

```python
def smart_division(func):
    def wrapper(a, b):
        if b == 0:
            print("Cannot divide by zero")
        else:
            return func(a, b)
    return wrapper

@smart_division
def divide(a, b):
    return a / b

print(divide(10, 2))
```

# 6. Decorator Returning Value

```python
def trace(func):
    def wrapper(*args, **kwargs):
        print("Executing:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@trace
def add(a, b):
    return a + b

print(add(5, 3))
```

# 7. Decorator with *args and **kwargs

Useful when you don't know number of arguments.

```python
@trace
def greet(name, age, city):
    print(name, age, city)
```

# 8. Decorators with Parameters (Decorator Factory)

```python
def prefix(msg):
    def decorator(func):
        def wrapper():
            print(msg)
            func()
        return wrapper
    return decorator

@prefix("Welcome!")
def hello():
    print("Hello user")

hello()
```

# 9. Multiple Decorators

```python
@decorator1
@decorator2
def fun():
    pass
```

Execution order:

```python
fun ‚Üí decorator2 ‚Üí decorator1
```

# 10. Class-Based Decorators

```python
class MyDecorator:
    def __init__(self, func):
        self.func = func

    def __call__(self):
        print("Before call")
        self.func()
        print("After call")

@MyDecorator
def hello():
    print("Hello!")

hello()
```

# 11. Real-Life Use Cases

**‚úî Logging**

```python
def log(func):
    def wrapper(*args, **kwargs):
        print("Calling", func.__name__)
        return func(*args, **kwargs)
    return wrapper
```

‚úî **Authentication**

```python
def check_admin(func):
    def wrapper(user):
        if user != "admin":
            print("Access denied")
        else:
            func(user)
    return wrapper
```

# ‚úî Measuring Execution Time

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print("Time:", end - start)
    return wrapper
```

# 12. functools.wraps (Very Important)

Without wraps, metadata (function name, docstring) is lost.

```python
from functools import wraps

def decorator(func):
    @wraps(func)
    def wrapper():
        return func()
    return wrapper
```
