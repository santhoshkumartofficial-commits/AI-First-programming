# Python Strings

A string is a sequence of characters inside quotes.

```python
name = "Python"
msg = 'Hello World'
long_text = """This is
a multi-line
string."""
```

# Basic String Operations
## 1. Accessing Characters

Strings work like arrays (index starts from 0).

```python
text = "Python"
print(text[0])   # P
print(text[3])   # h
```

# 2. Slicing

Extract part of a string.

```python
text = "Python"
print(text[0:3])    # Pyt
print(text[2:])     # thon
print(text[:4])     # Pyth
```

# 3. String Length

```python
text = "Hello"
print(len(text))   # 5
```

# 4. Concatenation (Joining)

```python
a = "Hello"
b = "World"
print(a + " " + b)   # Hello World
```

# 5. Repetition

```python
print("Hi" * 3)   # HiHiHi
```

# String Methods (Most Useful)
## 1. upper() / lower()

```python
text = "Python"
print(text.upper())   # PYTHON
print(text.lower())   # python
```

# 2. strip()

Removes spaces at start & end.

```python
text = "  Hello  "
print(text.strip())    # Hello


1.   List item
2.   List item
```

# 3. replace()

```python
text = "I like Java"
print(text.replace("Java", "Python"))  # I like Python
```

# 4. split()

Converts string → list.

```python
text = "apple,banana,orange"
print(text.split(","))
# ['apple', 'banana', 'orange']
```

# 5. join()

List → String.

```python
items = ['a', 'b', 'c']
print("-".join(items))
# a-b-c
```

# 6. find()

Returns index of substring.

```python
text = "hello"
print(text.find("l"))   # 2
```

# 7. count()

```python
text = "banana"
print(text.count("a"))   # 3
```

# String Formatting
## 1. f-string (Best way)

```python
name = "Santhosh"
age = 21
print(f"My name is {name} and I am {age} years old.")
```

# 2. format()

```python
print("My name is {} and I am {} years old".format("John", 25))
```

# String Example Program

```python
text = "Python Programming"

print("Uppercase:", text.upper())
print("Length:", len(text))
print("First 6 letters:", text[:6])
print("Replaced:", text.replace("Python", "Java"))
```

## 1. isalpha() — Checks only letters

```python
print("abc".isalpha())
print("abc123".isalpha())
```

# 2. isdigit() — Checks only digits

```python
print("123".isdigit())
print("12a3".isdigit())
```

# 3. isalnum() — Checks letters + numbers (no spaces)

```python
print("abc123".isalnum())
print("abc 123".isalnum())
```

# 4. islower() — Checks all lowercase

```python
print("hello".islower())
print("Hello".islower())
```

# 5. isupper() — Checks all uppercase

```python
print("HELLO".isupper())
print("Hello".isupper())
```

# ⭐ Extra Helpful Methods (Bonus)
## istitle() — First letter capital in each word

```python
print("Hello World".istitle())
print("hello world".istitle())
```

# isspace() — Only spaces

```python
print("   ".isspace())
print(" abc ".isspace())
```
