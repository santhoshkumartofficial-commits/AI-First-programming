# **Python Data Type Examples**

```python
# 1. Numeric Types
print("--- 1. NUMERIC TYPES ---")
age = 25                  # int
price = 99.99             # float
complex_num = 1 + 2j      # complex

print(f"age: {age}, Type: {type(age)}")
print(f"price: {price}, Type: {type(price)}")
print(f"complex_num: {complex_num}, Type: {type(complex_num)}")

# 2. Boolean Type
print("\n--- 2. BOOLEAN TYPE ---")
is_active = True          # bool
is_empty = False          # bool

print(f"is_active: {is_active}, Type: {type(is_active)}")

# 3. String Type
print("\n--- 3. STRING TYPE ---")
greeting = "Hello Python"  # str
multi_line_str = '''A long comment
spanning multiple lines.''' # str (used for multi-line text)

print(f"greeting: '{greeting}', Type: {type(greeting)}")
print(f"multi_line_str: '{multi_line_str}', Type: {type(multi_line_str)}")


# 4. Collection/Container Types

## 4a. List (Mutable)
print("\n--- 4a. LIST (MUTABLE) ---")
fruits = ["apple", "banana", "cherry"]  # list

# List is mutable: we can change it
fruits.append("orange")

print(f"fruits: {fruits}, Type: {type(fruits)}")


## 4b. Tuple (Immutable)
print("\n--- 4b. TUPLE (IMMUTABLE) ---")
coordinates = (10.0, 20.0, 30.0) # tuple

# Note: Attempting to modify a tuple (e.g., coordinates.append(40))
# would result in an error (AttributeError).

print(f"coordinates: {coordinates}, Type: {type(coordinates)}")


## 4c. Set (Mutable, Unordered, Unique)
print("\n--- 4c. SET (MUTABLE, UNIQUE) ---")
unique_numbers = {1, 2, 2, 3, 3, 4} # set (duplicates are automatically removed)
unique_numbers.add(5)

print(f"unique_numbers: {unique_numbers}, Type: {type(unique_numbers)}")


## 4d. Dictionary (Mutable Mapping)
print("\n--- 4d. DICTIONARY (MUTABLE MAPPING) ---")
person = {
    "name": "Alex",
    "age": 35,
    "city": "London"
} # dict

# Dictionary is mutable: we can add a new key-value pair
person["occupation"] = "Engineer"

print(f"person: {person}, Type: {type(person)}")


# 5. Special Type
print("\n--- 5. SPECIAL TYPE ---")
result = None   # NoneType

print(f"result: {result}, Type: {type(result)}")
```
