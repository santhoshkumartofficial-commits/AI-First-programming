# print stament
The print() statement is the fundamental function in Python used to display output to the console.

Here is a complete guide on how to use the print() statement, covering its basic use and key features.

1. Basic Syntax
The print() function in Python 3 is a built-in function that takes one or more objects as arguments and displays them on the screen.

# Simple output

To print a simple value, put the text or variable inside the parentheses:



```python
# Printing a string (text)
print("Hello, world!")

# Printing a number
print(42)

# Printing the value of a variable
name = "Alice"
print(name)

# Output:
# Hello, world!
# 42
# Alice
```

# Printing Multiple Items

You can pass multiple arguments to print(), separated by commas.

The sep (Separator) Parameter
By default, the print() function inserts a single space between each item. You can change this behavior using the optional sep parameter.

```python
day = 10
month = "July"
year = 2025

# Default: Separated by a space
print(day, month, year)

# Using a custom separator (e.g., a hyphen)
print(day, month, year, sep="-")

# Output:
# 10 July 2025
# 10-July-2025
```

# Ending the Output

By default, the print() function adds a newline character (\n) at the end of the output, meaning the next print() statement will start on a new line. You can change this using the optional end parameter.

The end Parameter
Setting the end parameter allows you to specify what should follow the printed content.



```python
# The first print statement will end with a space instead of a newline
print("Starting sequence...", end=" ")

# The second print statement continues on the same line
print("Complete.")

# Output:
# Starting sequence... Complete.
```

# Printing Variables and Strings Together

You have three common and highly recommended ways to combine variables and descriptive text in a print statement:

A. F-Strings (Formatted String Literals) - Recommended
F-strings (available since Python 3.6) are the cleanest and most modern way to format strings. You prefix the string with an f and put variables directly inside curly braces {}.

```python
item = "Laptop"
price = 1200

# Using an f-string
print(f"The {item} costs ${price}.")

# Output:
# The Laptop costs $1200.
```

# The Comma Separator


Using the comma automatically inserts a space, but can be less flexible than f-strings.

```python
x = 5
y = 10
print("The sum of", x, "and", y, "is", x + y)

# Output:
# The sum of 5 and 10 is 15
```

# The .format() Method (Older Standard)

This method is still widely used but is generally superseded by f-strings.

```python
print("The {} costs ${}.".format(item, price))

# Output:
# The Laptop costs $1200.
```

# Newline and Tab Characters
You can use escape sequences inside your string arguments to control formatting:

```python
# Using a newline character
print("First line.\nSecond line.")

# Output:
# First line.
# Second line.
```
