# üî• 4. How to Import a Module
**‚úî Method 1: import module**

```python
import math
print(math.sqrt(16))
```

**‚úî Method 2: import specific function**

```python
from math import sqrt
print(sqrt(16))
```

**‚úî Method 3: import with alias**

```python
import math as m
print(m.sqrt(25))
```

**‚úî Method 4: import everything (NOT recommended)**

```python
from math import *
```

# üî• 5. Creating Your Own Module (User-defined Module)
**Step 1: Create a file**

üìÅ mymath.py

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b
```

**Step 2: Import and use it**

üìÅ main.py

```python
import mymath

print(mymath.add(10, 20))
print(mymath.sub(50, 15))
```

#  8. The __name__ == "__main__" Concept

Prevents code from running when the module is imported.

```python
def show():
    print("Hello!")

if __name__ == "__main__":
    show()
```

# 9. Packages in Python

A package is a collection of modules inside a folder containing __init__.py.

```python
mypackage/
   __init__.py
   module1.py
   module2.py
```

Importing:

```python
from mypackage import module1
```
