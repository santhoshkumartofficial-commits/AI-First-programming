# ‚úÖ 1. Lambda Function (Anonymous Function)
**‚úî What is a lambda?**

A lambda is a small anonymous function defined without a name.

**‚úî Syntax:**

```python
lambda arguments: expression
```

**‚úî Example 1: Add 10 to a number**

```python
add10 = lambda x: x + 10
print(add10(5))   # Output: 15
```

**‚úî Example 2: Multiply two numbers**

```python
mul = lambda a, b: a * b
print(mul(3, 4))  # Output: 12
```

# **üìå Why use lambda?**

Short and simple functions

Used inside map(), filter(), reduce()

One-line functions

# ‚≠ê 2. map() Function

map() applies a function to every item in an iterable.

**‚úî Syntax:**

```python
map(function, iterable)
```

**‚úî Example 1: Square each number**

```python
nums = [1, 2, 3, 4]
result = list(map(lambda x: x*x, nums))
# Output: [1, 4, 9, 16]
```

**‚úî Example 2: Convert list of strings to uppercase**

```python
names = ["santhosh", "kumar"]
up = list(map(lambda s: s.upper(), names))
```

# üìå map() with multiple lists

```python
a = [1, 2, 3]
b = [10, 20, 30]
result = list(map(lambda x, y: x + y, a, b))
# Output: [11, 22, 33]
```

# ‚≠ê 3. filter() Function

filter() keeps only the elements that satisfy a condition.

**‚úî Syntax:**

```python
filter(function, iterable)
```

**‚úî Example 1: Keep only even numbers**

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
# Output: [2, 4, 6]
```

 ** ‚úî Example 2: Filter names starting with 'A'**

```python
names = ["Abi", "Bala", "Anu", "Vijay"]
result = list(filter(lambda x: x.startswith("A"), names))
# Output: ['Abi', 'Anu']
```

# ‚≠ê 4. reduce() Function

reduce() applies a function to elements and reduces them to a single value.

**üìå reduce() is in the functools module**

```python
from functools import reduce
reduce(function, iterable)
```

**‚úî Example 1: Sum of list**

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda x, y: x + y, nums)
# Output: 15
```

**‚úî Example 2: Find maximum value**

```python
largest = reduce(lambda a, b: a if a > b else b, nums)
```

**‚úî Example 3: Multiply all numbers**

```python
product = reduce(lambda x, y: x * y, nums)
```

| Function   | Purpose                           | Output                   |
| ---------- | --------------------------------- | ------------------------ |
| `map()`    | Apply function to each item       | List (transformed items) |
| `filter()` | Keep items that satisfy condition | List (filtered items)    |
| `reduce()` | Reduce list to single value       | One result               |


# üî• Combined Example (map + filter + lambda)
**Square only even numbers**

```python
nums = [1, 2, 3, 4, 5, 6]

result = list(
    map(lambda x: x*x,
        filter(lambda x: x % 2 == 0, nums))
)
# Output: [4, 16, 36]
```
