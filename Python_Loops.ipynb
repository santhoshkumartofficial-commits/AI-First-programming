# 1. FOR LOOP

Used when number of iterations is known.

Basic Syntax

```python
for variable in iterable:
    statement
```

# ‚úîÔ∏è Example 1: Loop through a list

```python
fruits = ["apple", "banana", "orange"]
for f in fruits:
    print(f)
```

# ‚úîÔ∏è Example 2: Loop through a string

```python
for ch in "hello":
    print(ch)
```

# ‚úîÔ∏è Example 3: Using range()

```python
for i in range(1, 6):
    print(i)
```

# üüß 2. WHILE LOOP

Used when number of iterations is NOT known (condition-based).

Basic Syntax

```python
while condition:
    statement
```

# ‚úîÔ∏è Example 1: Print 1 to 5

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

# ‚úîÔ∏è Example 2: Reverse loop

```python
i = 5
while i > 0:
    print(i)
    i -= 1
```

# 3. Loop Control Statements

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

üîπ continue ‚Äì skip current iteration

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

# üîπ pass ‚Äì empty statement (do nothing)

```python
for i in range(5):
    pass   # used as a placeholder
```

# 4. Looping Through Dictionaries
‚úîÔ∏è Print keys

```python
for key in d.keys():
    print(key)
```

‚úîÔ∏è Print values

```python
for val in d.values():
    print(val)
```

‚úîÔ∏è Print key + value

```python
for k, v in d.items():
    print(k, v)
```

# ‚≠ê 5. Patterns Using Loops
## ‚≠ê Pattern 1: Right Triangle

```python
*
**
***
****
```

```python
for i in range(1, 5):
    print("*" * i)
```

# ‚≠ê Pattern 2: Number Pyramid

```python
1
12
123
1234

```

```python
for i in range(1, 5):
    for j in range(1, i+1):
        print(j, end="")
    print()
```
