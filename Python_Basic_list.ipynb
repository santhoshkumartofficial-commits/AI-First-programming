# üî¢ Indexing in Lists

Python indexing starts from 0.

| Index | 0     | 1      | 2      |
| ----- | ----- | ------ | ------ |
| Item  | apple | banana | cherry |


# Accessing elements

```python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])   # apple
print(fruits[1])   # banana
print(fruits[2])   # cherry
```

# Negative Indexing

| Index | -3    | -2     | -1     |
| ----- | ----- | ------ | ------ |
| Item  | apple | banana | cherry |


```python
print(fruits[-1])   # cherry
print(fruits[-2])   # banana
```

# List Slicing (Getting a part of the list)

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])   # [20, 30, 40]
print(numbers[:3])    # [10, 20, 30]
print(numbers[2:])    # [30, 40, 50]
```

# Common List Operations
## 1. Add items

```python
fruits.append("orange")
```

# 2. Insert at specific position

```python
fruits.insert(1, "mango")
```

# 3. Remove items

```python
fruits.remove("banana")
```

# 4. Delete by index

```python
del fruits[2]
```

# 5. Replace item

```python
fruits[0] = "grape"
```

# 6. Length of list

```python
len(fruits)
```

# Looping through a list

```python
for item in fruits:
    print(item)
```

```python
fruits = ["apple", "banana", "cherry", "orange"]

print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])

print("Middle items:", fruits[1:3])

fruits.append("grape")
print("Updated list:", fruits)
```

| **Method**  | **Description**                       | **Example**                          |
| ----------- | ------------------------------------- | ------------------------------------ |
| `append()`  | Adds item to end                      | `fruits.append("mango")`             |
| `insert()`  | Adds item at specific index           | `fruits.insert(1, "kiwi")`           |
| `extend()`  | Adds multiple items                   | `fruits.extend(["grape", "orange"])` |
| `remove()`  | Removes first occurrence of an item   | `fruits.remove("apple")`             |
| `pop()`     | Removes item at index (default: last) | `fruits.pop(1)`                      |
| `clear()`   | Removes all items from list           | `fruits.clear()`                     |
| `index()`   | Returns index of an item              | `fruits.index("banana")`             |
| `count()`   | Counts occurrences of an item         | `nums.count(2)`                      |
| `sort()`    | Sorts list (ascending by default)     | `nums.sort()`                        |
| `reverse()` | Reverses list order                   | `nums.reverse()`                     |
| `copy()`    | Returns a copy of list                | `new_list = nums.copy()`             |
