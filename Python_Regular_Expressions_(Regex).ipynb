# Python Regular Expressions

# 1. What is Regex?

Regex (Regular Expressions) are patterns used to search, match, and manipulate text.

Used for:

Validating emails, phone numbers

Searching patterns

Extracting data

Replacing words

Log file analysis




# 2. Important Regex Functions (re module)


| Function       | Description             |
| -------------- | ----------------------- |
| `re.search()`  | Finds first match       |
| `re.findall()` | Returns all matches     |
| `re.match()`   | Match only at beginning |
| `re.sub()`     | Replace                 |
| `re.split()`   | Split by pattern        |
| `re.compile()` | Precompile pattern      |


# 3. Meta Characters (Very Important)

| Symbol  | Meaning                      |    |
| ------- | ---------------------------- | -- |
| `.`     | Any character except newline |    |
| `^`     | Start of string              |    |
| `$`     | End of string                |    |
| `*`     | 0 or more occurrences        |    |
| `+`     | 1 or more occurrences        |    |
| `?`     | 0 or 1 occurrence            |    |
| `{n}`   | Exactly n times              |    |
| `{n,m}` | n to m repetitions           |    |
| `[]`    | Character set                |    |
| `()`    | Grouping                     |    |
| `       | `                            | OR |


# 4. Special Sequences

| Pattern | Meaning               |
| ------- | --------------------- |
| `\d`    | Digit (0‚Äì9)           |
| `\D`    | Not digit             |
| `\w`    | Word char (a-z,0-9,_) |
| `\W`    | Not word char         |
| `\s`    | Space                 |
| `\S`    | Not space             |
| `\b`    | Word boundary         |
| `\B`    | Not boundary          |


# 5. Common Patterns (Cheat Sheet)


| Purpose           | Regex Pattern                         |
| ----------------- | ------------------------------------- |
| Email             | `\w+@\w+\.\w+`                        |
| Phone (10 digits) | `\d{10}`                              |
| Indian mobile     | `[6-9]\d{9}`                          |
| Postal code       | `\d{6}`                               |
| Only letters      | `^[A-Za-z]+$`                         |
| Only numbers      | `^\d+$`                               |
| Strong password   | `(?=.*\d)(?=.*[A-Z])(?=.*[a-z]).{8,}` |
| Date (dd-mm-yyyy) | `\d{2}-\d{2}-\d{4}`                   |


# 6. Basic Examples

Example 1: search()

```python
import re
txt = "I love Python"
print(re.search("Python", txt))
```

# Example 2: findall()


```python
re.findall("\d", "A1B2C3")
# Output: ['1','2','3']
```

# Example 3: split()

```python
re.split("\s", "Hello World Python")
# ['Hello','World','Python']
```

# Example 4: sub()

```python
re.sub("Python", "Java", "Python is easy")
# Java is easy
```

# 7. Character Sets

```python
re.findall("[aeiou]", "hello world")
# vowels = ['e','o','o']
```

Range:

```python
[A-Z], [a-z], [0-9]
```

8. Quantifiers
* ‚Üí 0 or more

```python
re.findall("ab*", "ab abbb a ac")
# ['ab', 'abbb', 'a']
```

? ‚Üí 0 or 1

```python
re.findall("ab?", "a ab abbb ac")
# ['a','ab','ab']
```

# 9. Grouping with ()

```python
re.findall("(ab)+", "abababab")
# ['ab','ab','ab','ab']
```

# 10. Compiling Patterns

Used for performance improvement.

```python
pattern = re.compile("\d{3}")
pattern.findall("123 456 789")
```

# 11. Real-Life Example: Extract Emails


```python
text = "Contact me at abc@gmail.com or test@yahoo.com"
emails = re.findall(r"\w+@\w+\.\w+", text)
print(emails)
```

# 12. Validate Phone Number

```python
pattern = r"^[6-9]\d{9}$"
print(bool(re.match(pattern, "9876543210")))
```

# 13. Regex Flags


| Flag   | Meaning            |
| ------ | ------------------ |
| `re.I` | Ignore case        |
| `re.M` | Multi-line         |
| `re.S` | Dot match new line |
| `re.X` | Verbose mode       |


```python
re.search("python", "PYTHON", re.I)
```
