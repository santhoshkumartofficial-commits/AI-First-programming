# üü® Python Dictionaries

A dictionary stores data in key‚Äìvalue pairs.
Written using { }

```python
student = {
    "name": "Santhosh",
    "age": 21,
    "course": "Python"
}
```

# üîë Dictionary Features

Keys must be unique

Values can be any data type

Order is maintained (Python 3.7+)

Changeable (mutable)

# üü¶ Accessing Values
## 1. Using the key

```python
print(student["name"])   # Santhosh
```

2. Using get()

```python
print(student.get("age"))   # 21
```

# üîÑ Changing Values

```python
student["age"] = 22
```

# ‚ûï Adding New Key‚ÄìValue Pair

```python
student["grade"] = "A"
```

# ‚ùå Removing Items
pop() ‚Äì removes by key

```python
student.pop("age")
```

# popitem() ‚Äì removes last item

```python
student.popitem()
```

# del ‚Äì delete by key


```python
del student["course"]
```

# clear() ‚Äì empty dictionary

```python
student.clear()
```

# üîÑ Dictionary Looping
# 1. Keys

```python
for key in student.keys():
    print(key)
```

# 2. Values

```python
for val in student.values():
    print(val)
```

# 3. Key & Value together




```python
for key, val in student.items():
    print(key, val)
```

# üß© Useful Dictionary Methods


| Method      | Description             | Example                      |
| ----------- | ----------------------- | ---------------------------- |
| `keys()`    | Returns all keys        | `student.keys()`             |
| `values()`  | Returns all values      | `student.values()`           |
| `items()`   | Returns key-value pairs | `student.items()`            |
| `get()`     | Safely access value     | `student.get("name")`        |
| `update()`  | Update multiple values  | `student.update({"age":22})` |
| `pop()`     | Remove item by key      | `student.pop("age")`         |
| `popitem()` | Remove last item        | `student.popitem()`          |
| `clear()`   | Empty dictionary        | `student.clear()`            |


# üß™ Example Program

```python
student = {
    "name": "Santhosh",
    "age": 21,
    "marks": [85, 90, 88]
}

print("Name:", student["name"])
print("Marks:", student["marks"])

student["age"] = 22
student["grade"] = "A"

for key, value in student.items():
    print(key, ":", value)
```

# üìå Python Dictionary Methods ‚Äî Full Table

| Method                         | Description                                            | Example                           |
| ------------------------------ | ------------------------------------------------------ | --------------------------------- |
| **`get(key, default)`**        | Returns value of key; returns default if key not found | `d.get("age", 0)`                 |
| **`keys()`**                   | Returns all keys                                       | `d.keys()`                        |
| **`values()`**                 | Returns all values                                     | `d.values()`                      |
| **`items()`**                  | Returns key‚Äìvalue pairs                                | `d.items()`                       |
| **`update(dict)`**             | Updates/adds key-value pairs                           | `d.update({"age": 25})`           |
| **`pop(key)`**                 | Removes item with key                                  | `d.pop("name")`                   |
| **`popitem()`**                | Removes **last** inserted item                         | `d.popitem()`                     |
| **`clear()`**                  | Removes all items                                      | `d.clear()`                       |
| **`setdefault(key, default)`** | Returns value; if key missing, adds key with default   | `d.setdefault("city", "Chennai")` |
| **`copy()`**                   | Returns a shallow copy                                 | `d2 = d.copy()`                   |
| **`fromkeys(keys, value)`**    | Creates a new dict with keys and one value             | `dict.fromkeys(["a","b"], 0)`     |
