# üìå 1. What is List Comprehension?

List Comprehension is a short and fast way to create lists using a single line of code.

‚úî Instead of using loops

‚úî More readable

‚úî Faster execution

# üìå 2. Syntax

```python
[ expression  for item in iterable  if condition ]
```

Parts:

expression ‚Üí What to store in the list

item ‚Üí Each element in the loop

iterable ‚Üí list, string, range, etc.

if condition (optional) ‚Üí Filter elements

# üìå 3. Basic Examples
üéØ Example 1: Create a list of numbers from 0‚Äì9

```python
nums = [x for x in range(10)]
```

üéØ Example 2: Squares of numbers

```python
squares = [x*x for x in range(1, 11)]
```

üéØ Example 3: Even numbers

```python
evens = [x for x in range(20) if x % 2 == 0]
```

üéØ Example 4: Convert strings to upper case

```python
names = ["santhosh", "kumar"]
upper_names = [name.upper() for name in names]
```

# üìå 4. List Comprehension with Condition (if‚Äìelse)

üéØ If‚ÄìElse inside expression

```python
result = ["Even" if x%2==0 else "Odd" for x in range(10)]
```

üìå 5. Using Multiple Conditions

```python
filtered = [x for x in range(50) if x % 2 == 0 if x > 10]
```

```python
üìå 6. Nested List Comprehension

Example: 2D Matrix
```

```python
matrix = [[j for j in range(3)] for i in range(3)]
```

Flatten a 2D list

```python
flat = [num for row in matrix for num in row]
```

# üìå 7. List Comprehension with Functions

```python
def square(n):
    return n*n

out = [square(x) for x in range(5)]
```

# üìå 8. List Comprehension with Strings

```python
vowels = [ch for ch in "santhoshkumar" if ch in "aeiou"]
```

# üìå 9. List Comprehension with Dictionaries
Get keys only

```python
keys = [key for key in mydict]
```

Convert dict values to list

```python
values = [v for v in mydict.values()]
```

# üìå 10. Using List Comprehension with Conditions + Functions

```python
result = [x*x for x in range(50) if x%5==0]
```

# üìå 11. Advanced Examples

üéØ Cartesian Product

```python
pairs = [(x, y) for x in [1,2,3] for y in [4,5]]
```

üéØ Numbers divisible by 3 but not 5

```python
nums = [x for x in range(50) if x%3==0 and x%5!=0]
```

üéØ Reverse each string in list

```python
rev = [s[::-1] for s in ["apple", "ball", "cat"]]
```

üéØ Extract digits from a string

```python
digits = [int(c) for c in "a1b2c3" if c.isdigit()]
```
