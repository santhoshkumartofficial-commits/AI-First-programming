# üî• 5. Adding Attributes to a Class

```python
class Student:
    name = "Santhosh"
    age = 20

s = Student()
print(s.name)
print(s.age)
```

# üî• 6. Using Constructor (init)

**__init__() **is automatically called when an object is created.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Arun", 21)
s2 = Student("Bala", 20)

print(s1.name, s1.age)
print(s2.name, s2.age)
```

# 7. Adding Methods to a Class

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print("Name:", self.name, "Age:", self.age)

s = Student("Kumar", 22)
s.show()
```

# üî• 12. Special Methods (Magic Methods)

Python supports magic methods like:

| Method     | Purpose               |
| ---------- | --------------------- |
| `__init__` | Constructor           |
| `__str__`  | String representation |
| `__len__`  | Length                |
| `__add__`  | Operator overloading  |


```python
class A:
    def __str__(self):
        return "This is object of class A"

a = A()
print(a)
```

# 13. Example: Real-Life Class (Car)

```python
class Car:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    def info(self):
        print(self.brand, self.model, "-", self.price)

c1 = Car("BMW", "X5", 8500000)
c2 = Car("Audi", "Q7", 9200000)

c1.info()
c2.info()
```
